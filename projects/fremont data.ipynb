```python
by_time = d.groupby(d.index.time).sum()
by_time.head()
```

```python
import numpy as np
plt.figure(figsize=(16,9))
hourly_ticks = 4*60*60 * np.arange(6)
by_time.plot(xticks=hourly_ticks,style=[':','--','-'])
```

```python
by_weekday = d.groupby(d.index.dayofweek).sum()
by_weekday.head()
```

```python
by_weekday.index= ['mon','tue','wed','thur','fri','sat','sun']
by_weekday.plot(style=[':','--','-'])
```

```python
by_monthly = d.groupby(d.index.month).sum()
by_monthly.head()
```

```python
by_monthly.index = ['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']
by_monthly.plot(style=[':','--','--'])
```

```python
weekend = np.where(d.index.weekday < 5,'weekday','weekend')
weekend[0:50]
```

```python
by_time = d.groupby([weekend,d.index.time]).sum()
by_time.head()
```

```python
by_time.tail()
```

```python
fig,ax = plt.subplots(1,2, figsize=(16,8))
by_time.loc['weekday'].plot(ax=ax[0], title='weekdays',
                            xticks= hourly_ticks,style=[':','--','-'])
by_time.loc['weekend'].plot(ax=ax[1],title='weekends',
                           xticks=hourly_ticks,style=[':','--','-'])
```
