```python
rng = np.random.RandomState()
df = pd.DataFrame({'key':['a','b','c','a','b','c'],
                'data1':range(6),
                'data2':rng.randint(0,10,0)},
                columns=['key','data1','data2'])
```

```python
df.groupby('key'.aggregate(['min',np.median,'max',np.mean]))
```

```python
df.agg(['min','max'])
```

```python
df.groupby('key').aggregate({'data1':'min',
                            'data2':'max'})
```

```python
def filter_func(x):
    return x['data2'].std()>4
```

```python
df.groupby('key').std()
```

```python
df.groupby('key').filter(filter_func)
```

# transfromations

```python
df.groupny('key').tansform(lamda x: x-x.mean())
```

```python
df.groupny('key').tansform(lamda x: x+x.mean())
```

# the apply() method

```python
def norm_by_data2(x):
    x['data1'] /= x['data2'].sum() #a/=b-->a=a/b
    return x
```

```python
df.groupby('key').apply(norm_by_data2)
```
